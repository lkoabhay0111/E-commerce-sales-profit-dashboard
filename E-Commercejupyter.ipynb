```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("C:/Users/LENOVO PC/Downloads/Sample - Superstore.csv",encoding="latin-1")
df=pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df["Order Date"]=pd.to_datetime(df["Order Date"],format="mixed")
df["Ship Date"]=pd.to_datetime(df["Ship Date"],format="mixed")
df["Order Date"]
df["Ship Date"]
```

```python
df["Month"]=df["Order Date"].dt.month_name()
df["Month"]
df
```

```python
#You need to calculate the monthly sales of the store and identify which month had the highest sales and which month had the lowest sales.
Monthly_sales=df.groupby("Month")["Sales"].sum().reset_index()
highest=Monthly_sales.nlargest(1,"Sales")
lowest=Monthly_sales.nsmallest(1,"Sales")
highest_month=highest["Month"].values[0]
highest_value=highest["Sales"].values[0]
lowest_month=lowest["Month"].values[0]
lowest_value=lowest["Sales"].values[0]
print("Highest_sale_month",highest_month,highest_value)
print("Lowest_sale_month",lowest_month,lowest_value)
```

```python
plt.figure(figsize=(12,5))
color=Monthly_sales["Sales"].apply(lambda x:"green" if x==highest_value else
                                   "red" if x==lowest_value else
                                   "blue")
plt.bar(Monthly_sales["Month"],Monthly_sales["Sales"],color=color)
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
#You need to analyze sales based on product categories and determine which category has the lowest sales and which category has the highest sales.
product_categories=df.groupby("Category")["Sales"].sum().reset_index()
highest=product_categories.nlargest(1,"Sales")
lowest=product_categories.nsmallest(1,"Sales")
Lowest_sale=lowest["Category"].values[0]
Lowest_value=lowest["Sales"].values[0]
Highest_sale=highest["Category"].values[0]
Highest_value=highest["Sales"].values[0]
print("Lowest_sale",Lowest_sale,Lowest_value)
print("Highest_sale",Highest_sale,Highest_value)
```

```python
color=product_categories["Sales"].apply(lambda x:"green" if x==Highest_value else
                                        "red" if x==Lowest_value else
                                        "blue")
                                         
plt.bar( product_categories["Category"],product_categories["Sales"],color=color)
plt.title("Sales by Product Category (Highest & Lowest Highlighted)")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
#The sales analysis need to be done based on sub-category
sale_sub_category=df.groupby("Sub-Category")["Sales"].sum().reset_index()
plt.bar(sale_sub_category["Sub-Category"],sale_sub_category["Sales"])
plt.xticks(rotation=90)
plt.xlabel("Sub-Category")
plt.ylabel("Total Sales")
plt.title("Sales by Sub-Category")
plt.show()
        
```

```python
#You need to analyze the monthly profit from sales and determine which month had the highest profit.
Monthly_profit=df.groupby("Month")["Profit"].sum().reset_index()
top_row=Monthly_profit.nlargest(1,"Profit")
highest_profit=top_row["Month"].values[0]
highest_profit_value=top_row["Profit"].values[0]
print("highest profit",highest_profit,highest_profit_value)
plt.bar(Monthly_profit["Month"],Monthly_profit["Profit"])
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.title("Monthly_profit")
plt.xticks(rotation=90)
plt.show()
```

 Profit by Category 


```python
category_profit = ( df.groupby("Category")["Profit"].sum().reset_index())
category_profit
plt.pie(category_profit["Profit"],labels=category_profit["Category"],autopct="%1.1f%%",startangle=90)
plt.title("Profit Distribution by Category")
plt.show()
```

 Profit by Sub-Category

```python
sub_category_profit =(df.groupby("Sub-Category")["Profit"].sum().reset_index())
sub_category_profit
plt.bar(sub_category_profit["Sub-Category"],sub_category_profit["Profit"])
plt.xlabel("Sub-Category")
plt.ylabel("Total Profit")
plt.title("Profit by Sub-Category")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#Analyze the sales and profit by customer segment.
sales_profit_by_segment=df.groupby("Segment").agg({"Sales":"sum","Profit":"sum"}).reset_index()
sales_profit_by_segment
```

```python
x = np.arange(len(sales_profit_by_segment))
width = 0.35

plt.bar(x - width/2, sales_profit_by_segment["Sales"],
        width, label="Sales")
plt.bar(x + width/2, sales_profit_by_segment["Profit"],
        width, label="Profit")

plt.xticks(x, sales_profit_by_segment["Segment"])
plt.xlabel("Customer Segment")
plt.ylabel("Amount")
plt.title("Sales and Profit by Customer Segment")
plt.legend()
plt.show()
```

```python
#Analyze the sales-to-profit ratio.
# Total sales and total profit
total_sales = df["Sales"].sum()
total_profit = df["Profit"].sum()

sales_to_profit_ratio = total_profit / total_sales

print("Total Sales:", round(total_sales, 2))
print("Total Profit:", round(total_profit, 2))
print("Sales-to-Profit Ratio:", round(sales_to_profit_ratio, 4))
```

```python

plt.bar(["Total Sales", "Total Profit"], [total_sales, total_profit])
plt.ylabel("Amount")
plt.title("Total Sales vs Total Profit")
plt.show()

```
